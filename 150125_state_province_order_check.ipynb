```python
a_sum = a_sum.rename(columns={"orders": "Alberta Orders"})
a_sum.head()
```

```python
# terst
```

```python
bc_sum = bc_sum.rename(columns={"orders": "British Columbia Orders"})
bc_sum.head()
```

```python
total = all_sum.merge(a_sum, how="left", left_on="p_date", right_on="p_date").merge(bc_sum, how="left", left_on="p_date", right_on="p_date")
```

```python
total = total.fillna(0)
```

```python
total.describe()
```

```python
total.head(1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 示例数据创建（实际使用时替换为你的数据）

df = total

plt.figure(figsize=(12, 6))

plt.plot(df['p_date'], df['orders'],
         marker='o', linestyle='-', color='royalblue', label='total')
plt.plot(df['p_date'], df['Alberta Orders'],
         marker='s', linestyle='--', color='crimson', label='Alberta Orders')
plt.plot(df['p_date'], df['British Columbia Orders'],
         marker='^', linestyle='-.', color='forestgreen', label='British Columbia Orders')


plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.gcf().autofmt_xdate()


plt.xlabel('Date', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.title('Daily Orders Trend ', fontsize=14, pad=20)
plt.legend(title='Province', fontsize=10)


plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
